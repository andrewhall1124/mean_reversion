```python
# Import dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#get some data
df = pd.read_parquet("../data/BOQ_data.parquet")

df
```

```python
# Sort out labels
labels = sorted(df.columns.to_list())

labels
```

```python
# Clea up the data
clean = df[['DATE', 'PERMNO', 'PRC', 'RET', 'ME']]

clean = clean.rename(columns={'DATE':'date', 'PERMNO': 'permno', 'PRC':'price', 'RET': 'return', 'ME': 'market_cap'})

clean
```

```python
# Ensure 'date' is in datetime format
clean['date'] = pd.to_datetime(clean['date'])

# Set 'date' as the index for easier time-based operations
clean.set_index('date', inplace=True)

# Define the cutoff period (2 years from the most recent date in the dataset)
two_years_ago = clean.index.max() - pd.DateOffset(years=2)

# Filter the dataset to keep only rows within the last 2 years
clean = clean[clean.index >= two_years_ago]

# Create a copy of the filtered DataFrame to avoid SettingWithCopyWarning
clean = clean.copy()

# Define the window size for the weighted moving average (3 months of data)
window_size = 3

# Function to calculate a weighted moving average with more recent data receiving higher weights
def weighted_moving_average(values):
    weights = np.arange(1, len(values) + 1)  # Increasing weights for more recent values
    return np.dot(values, weights) / weights.sum()

# Apply the weighted moving average to the 'price' column using .loc for safe assignment
clean.loc[:, 'weighted_moving_avg'] = clean['price'].rolling(window=window_size, min_periods=1).apply(weighted_moving_average, raw=True)

# Display the last 10 rows of the DataFrame with the original 'price' and calculated 'weighted_moving_avg'
print(clean[['price', 'weighted_moving_avg']].tail())

```
